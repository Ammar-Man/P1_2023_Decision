```python
# Dataset från youtube users 
import pandas as pd
df = pd.read_csv('./USvideos.csv')
df.columns
```

```python
#  utforska och analysera ert data set (EDA)
print(df.head())  # Visar de första raderna av datasetet
print(df.info())  # Visar information om datatyper och null-värden
print(df.describe())  # Visar statistiska detaljer om numeriska kolumner
print(df['tags'].value_counts())  # Utforskar unika värden och deras förekomst

```

### Bygga Ett Rekommendationssystem Baserat på Innehållsbaserad Filtrering:

```python
# Förbereder Data:  tekniker som TF-IDF för att representera texten numeriskt.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

tfidf_vectorizer = TfidfVectorizer()  # Skapar en TF-IDF Vectorizer
tfidf_matrix = tfidf_vectorizer.fit_transform(df['title'])  # Omvandlar texten till TF-IDF-matris

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)  # Beräknar kosinuslikheten mellan objekten
```

```python
#  Rekommendationer: Definiera en funktion som tar in en title
def get_recommendations(title):
    idx = df[df['title'] == title].index[0]  # Få index för den valda title
    sim_scores = list(enumerate(cosine_sim[idx]))  # Hämta likheter från cosine_sim-matrisen
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)  # Sortera objekten baserat på likheter
    similar_articles = sim_scores[1:11]  # Få de 10 mest liknande objekten
    return similar_articles
get_recommendations("WE WANT TO TALK ABOUT OUR MARRIAGE")
```

```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import KNNBasic

reader = Reader(rating_scale=(0, 5.613827e+06))  # Ange skalan för betyg
data = Dataset.load_from_df(df[['category_id', 'likes','dislikes' ]], reader)
trainset, testset = train_test_split(data, test_size=0.2)  # Dela upp data i träning och test

sim_options = {
    'name': 'cosine',  # Välj den kollaborativa filtreringstekniken
    'user_based': False  # Använd item-baserad filtrering
}
model = KNNBasic(sim_options=sim_options)  # Skapa modell
model.fit(trainset)  # Träna modellen
```

```python
channel_title = "CaseyNeistat"
user_id = channel_title

category_id = "22"
item_ids =category_id


item_ids = df['category_id'].unique()  # Hämta unika objekt från datasetet
predicted_ratings = [model.predict(user_id, item_id).est for item_id in item_ids]  # Få förutsagda betyg för objekten
recommendations = pd.DataFrame({'category_id': item_ids, 'predicted_rating': predicted_ratings})
recommendations = recommendations.sort_values(by='predicted_rating', ascending=False).head(10)  # Få de 10 bästa rekommendationerna
recommendations
```
